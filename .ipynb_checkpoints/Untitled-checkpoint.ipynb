Membuat list "data" dan dataframe "datas" yang kosong. dataframe datas akan digunakan untuk menyimpan data peserta yang mempunyai penyakit F20|F80|F41|F32|F25|F31|F45|F23|F06|F03

```python
data1 = pd.read_csv('../data/01_kepesertaan.txt', sep="|", header=None)
```

Load dataset 1

```python
for i in f:
    datas=datas.append(data1.loc[data1[0]==i])
```

Pada proses ini dilakukan pencarian data pada dataset1, dimana data yang no_pesertanya ada pada list f akan dimasukkan kedalam dataframe datas

```python
datas = datas.drop_duplicates()
```

drop data duplicate data dataframe datas

```python
datas=datas.reset_index(drop=True)
```

mereset index pada dataframe datas

```python
penyakit=penyakit.rename(columns={1:'kode_penyakit'})
```

merename columns pada dataframe penyakit agar data mudah dibaca

```python
datas=datas.rename(columns={0:"no_peserta"})
penyakit=penyakit.rename(columns={0:"no_peserta"})
```

merename columns pada dataframe penyakit dan datas agar data mudah dibaca dan agar proses inner join mudah dilakukan

```python
penyakit = penyakit.drop_duplicates()
```

```python
result=pd.DataFrame()
result = pd.merge(datas, penyakit, left_on='no_peserta', right_on='no_peserta', how='inner')
result = result.drop_duplicates()
```

proses inner join pada dataframe datas dan penyakit guna mendapatkan kode penyakit

```python
result=result.rename(columns={1:"no_keluarga", 2:"tanggal_lahir", 3:"hubungan_keluarga", 4:"jenis_kelamin", 9:"kota", 5:"status_perkawinan"})
```

```python
result_copy=result
result_copy=result_copy.drop(columns=[6,7,8,10,11,12,13,14,15,16,17])
```

```python
result_copy=result_copy.rename(columns={'tanggal_lahir': 'umur'})
```
